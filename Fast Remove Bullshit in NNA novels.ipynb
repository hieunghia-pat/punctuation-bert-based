```python
import re
import os

for file_name in os.listdir('Raw_data_NNA/'):
    if file_name == 'clear': continue
    with open(f'Raw_data_NNA/clear/{file_name}', 'w', encoding='utf') as out_file:
        with open(f'Raw_data_NNA/{file_name}', 'r', encoding='utf') as in_file:
            for line in in_file:
                skip_line = False
                if 'oOo' == line.strip():
                    skip_line = True
                if 'N.N.A.' == line.strip():
                    skip_line = True
                if '*' in line.strip():
                    skip_line = True
                    print(line)
                if line.strip().isupper():
                    print(line)
                    skip_line = True


                if "Nguyễn Nhật Ánh" in line:
                    print(line)
                    skip_line = True
                if 'http' in line or 'www.' in line:
                    print(line)
                    skip_line = True
                if line.strip().isnumeric():
                    skip_line = True
                    print("Num:", line)
                if (re.search(r'^\d+', line.strip()) is not None  and line.strip().count(' ') < 2) or re.search(r'^chương \d+', line.strip(), flags=re.I) is not None:
                    print(line)
                    skip_line = True

                if len(line.strip()) >= 1 and line.strip().count(' ') < 1:
    #                 skip_line = True
                    print('Short sentence:', line)
                if not skip_line:
                    out_file.write(line)
```
